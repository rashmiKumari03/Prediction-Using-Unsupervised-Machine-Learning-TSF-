# How do you find the optimum number of clusters for K Means? How does one determine the value of K?

example:

```python
# Finding the optimum number of clusters for k-means classification

x = iris.iloc[:, [0, 1, 2, 3]].values
x
```

```python
from sklearn.cluster import KMeans

Centroid_list = []
No_of_clustering=range(1, 11)

for i in No_of_clustering :
    kmeans = KMeans(n_clusters = i, init = 'k-means++',max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    Centroid_list.append(kmeans.inertia_)

    
plt.plot(No_of_clustering,Centroid_list)
plt.show()
```

```python
# Plotting the results onto a line graph, 
# `allowing us to observe 'The elbow'
plt.plot(range(1, 11),Centroid_list)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Centroid_list') # Within cluster sum of squares
plt.show()
```

# Training and Prediction

```python
# Applying kmeans to the dataset / Creating the kmeans classifier


kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
#Predicting the value 
y_kmeans = kmeans.fit_predict(x)
```

```python
#Plotting The predicted values of Setosa.
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1],s = 100, c = 'red', label = 'Iris-setosa')
plt.legend()
```

```python
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1],s = 100, c = 'blue', label = 'Iris-versicolour')
plt.legend()
```

```python
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],s = 100, c = 'green', label = 'Iris-virginica')
plt.legend()
```

```python
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1],s = 100, c = 'yellow', label = 'Centroids')
plt.legend()
```

# Combing all the scatter plots of iris flower type and centroid.

```python
# Visualising the clusters - On the first two columns
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1],s = 100, c = 'pink', label = 'Iris-setosa')

plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1],s = 100, c = 'orange', label = 'Iris-versicolour')

plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],s = 100, c = 'green', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1],s = 100, c = 'blue', label = 'Centroids')

plt.legend()
```
